```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
scores = Path("output_data/scores.csv")

score_data = pd.read_csv(scores)

score_data
```

```python
# Filter the data to only show the required dates
score_data_filtered = score_data[~score_data['schedule_season'].between(1966, 2012)]

score_data_filtered
```

```python
# Define the mapping of names to be changed
name_mapping = {
    'Arizona Cardinals': 'ARI',
    'Atlanta Falcons': 'ATL',
    'Baltimore Ravens':'BAL',
    'Buffalo Bills':'BUF',
    'Carolina Panthers':'CAR',
    'Chicago Bears':'CHI',
    'Cincinnati Bengals':'CIN',
    'Cleveland Browns':'CLE',
    'Dallas Cowboys':'DAL',
    'Denver Broncos':'DEN',
    'Detroit Lions':'DET',
    'Green Bay Packers':'GB',
    'Houston Texans':'HOU',
    'Indianapolis Colts':'IND',
    'Jacksonville Jaguars':'JAX',
    'Kansas City Chiefs':'KC',
    'Las Vegas Raiders':'LVR',
    'Los Angeles Chargers':'LAC',
    'Los Angeles Rams':'LAR',
    'Miami Dolphins':'MIA',
    'Minnesota Vikings':'MIN',
    'New England Patriots':'NE',
    'New Orleans Saints':'NO',
    'New York Giants':'NYG',
    'New York Jets':'NYJ',
    'Philadelphia Eagles':'PHI',
    'Pittsburgh Steelers':'PIT',
    'San Francisco 49ers':'SF',
    'Seattle Seahawks':'SEA',
    'Tampa Bay Buccaneers':'TB',
    'Tennessee Titans':'TEN',
    'Washington Commanders':'WAS',
    'Oakland Raiders':'LVR',
    'St. Louis Rams':'LAR',
    'Washington Football Team':'WAS',
    'Washington Redskins':'WAS',
    'San Diego Chargers':'LAC'
}

# Replace the names in the "team_home" and "team_away" columns
score_data_filtered['team_home'] = score_data_filtered['team_home'].replace(name_mapping)
score_data_filtered['team_away'] = score_data_filtered['team_away'].replace(name_mapping)

# Display the updated DataFrame
score_data_filtered
```

```python
# Export the score_data_filtered into a csv
score_data_filtered.to_csv("output_data/nfl.csv")
```
